```python
import pytesseract
from PIL import Image, UnidentifiedImageError
import os
import csv

# Specify the path to the Tesseract executable
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Specify the path to your image directory
image_dir = "C:\\Users\\varch\\Downloads\\Imagesforproject"
i=0;

# Open the CSV file in write mode
with open('extracted_text.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    # Write the header
    writer.writerow(["File Name", "Extracted Text"])

    # Loop through all the images in the directory
    for image_file in os.listdir(image_dir):
        # Only process .jpg files
        if image_file.endswith(""):
            i+=1;
            # Construct the full image path
            full_image_path = os.path.join(image_dir, image_file)

            # Open the image file
            try:
                with Image.open(full_image_path) as img:
                    # Use pytesseract to extract text from the image
                    extracted_information = pytesseract.image_to_string(img)
                    # Replace newline characters with spaces
                    extracted_information = extracted_information.replace('\n', ' ')
                # Write the extracted text to the CSV file
                writer.writerow([image_file, extracted_information])
                print(f"Extracted text from {image_file} and wrote to CSV. File Number: {i}\n")
            except UnidentifiedImageError:
                print(f"Cannot identify image file {image_file}. Skipping...\n")
```
