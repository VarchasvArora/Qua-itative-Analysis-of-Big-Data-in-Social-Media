```python
from google_images_search import GoogleImagesSearch
import zipfile
import os
import re

# you can provide API key and CX using arguments,
# or you can set environment variables: GCS_DEVELOPER_KEY, GCS_CX
gis = GoogleImagesSearch('API key' , 'CX' , validate_images=True)

def safe_file_name(file_name):
    # Remove invalid characters
    safe_name = re.sub(r'[\\/*?:"<>|]', "", file_name)
    return safe_name

def search(keyword, imageNumber):
    _search_params = {
        'q': keyword,
        'num': imageNumber,
        # 'safe': 'medium',
        # 'fileType': 'jpg',
        # 'imgType': 'photo',
        # 'imgSize': 'MEDIUM',
        # 'imgDominantColor': 'brown',
        # 'rights': 'cc_publicdomain'
    }

    #path_to_dir: where the downloaded images must be stored
    gis.search(search_params=_search_params, path_to_dir='C:\\Users\\varch\\Downloads\\BDMImages')

    # Apply the safe_file_name function to all downloaded files
    for file_name in os.listdir('C:\\Users\\varch\\Downloads\\BDMImages'):
        safe_name = safe_file_name(file_name)
        os.rename(os.path.join('C:\\Users\\varch\\Downloads\\BDMImages', file_name), 
                  os.path.join('C:\\Users\\varch\\Downloads\\BDMImages', safe_name))

##calling search function to download 10 motorbike images
search('Big Data Social Media',50)
```
